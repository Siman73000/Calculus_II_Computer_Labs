<div style="font-size:20px;margin-bottom:10px;color:#7dd3fc;">Calculus of Gabriel's Horn</div>
<div style="color:#94a3b8;font-size:13px;margin-bottom:18px;">Paradox of finite volume and infinite surface area</div>

<!-- PART (a) -->
<div style="background:rgba(255,255,255,0.02);border:1px solid rgba(125,211,252,0.06);border-radius:12px;padding:20px;margin-bottom:24px;box-shadow:0 6px 30px rgba(2,6,23,0.6);">
  <div style="display:flex;align-items:center;gap:12px;margin-bottom:14px;font-weight:700;color:#e6f6ff;">
    <div style="width:10px;height:10px;background:#7dd3fc;border-radius:4px;"></div>
    <div>TASK (a) – Calculate the Volume for \(1 < x < \infty\)</div>
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    For the curve \(y = \frac{1}{x}\), the solid of revolution about the x-axis has volume:
    \[
      V = \pi \int_{1}^{\infty} y^2\,dx
      = \pi \int_{1}^{\infty} \frac{1}{x^2}\,dx.
    \]
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    Compute the improper integral:
    \[
      \pi \int_{1}^{\infty} x^{-2}\,dx
      = \pi \big[-x^{-1}\big]_{1}^{\infty}
      = \pi(0 - (-1)) = \boxed{\pi}.
    \]
  </div>
  <div style="color:#94a3b8;font-size:13px;">
    The volume of Gabriel’s Horn on \(1 < x < \infty\) is <strong>\(V = \pi\)</strong>.
  </div>
</div>

<!-- PART (b) -->
<div style="background:rgba(255,255,255,0.02);border:1px solid rgba(125,211,252,0.06);border-radius:12px;padding:20px;margin-bottom:24px;box-shadow:0 6px 30px rgba(2,6,23,0.6);">
  <div style="display:flex;align-items:center;gap:12px;margin-bottom:14px;font-weight:700;color:#e6f6ff;">
    <div style="width:10px;height:10px;background:#ffd166;border-radius:4px;"></div>
    <div>TASK (b) – Find the Antiderivative Representing Surface Area</div>
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    Surface area for revolution about the x-axis:
    \[
      S = 2\pi \int_{1}^{\infty} y \sqrt{1 + (y')^2}\,dx.
    \]
    For \(y = \frac{1}{x}\), we have \(y' = -\frac{1}{x^2}\), so:
    \[
      S = 2\pi \int_{1}^{\infty} \frac{1}{x} \sqrt{1 + \frac{1}{x^4}}\,dx.
    \]
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    The antiderivative for
    \(\displaystyle \int \frac{1}{x}\sqrt{1+\frac{1}{x^4}}\,dx\)
    is:
    \[
      \boxed{
        I(x) = \tfrac{1}{2}\operatorname{arsinh}(x^2)
        - \tfrac{1}{2}\sqrt{1+\frac{1}{x^4}} + C
      }.
    \]
  </div>
  <div style="color:#94a3b8;font-size:13px;">
    This expression gives the indefinite integral used for the surface area computation.
  </div>
</div>

<!-- PART (c) -->
<div style="background:rgba(255,255,255,0.02);border:1px solid rgba(125,211,252,0.06);border-radius:12px;padding:20px;margin-bottom:24px;box-shadow:0 6px 30px rgba(2,6,23,0.6);">
  <div style="display:flex;align-items:center;gap:12px;margin-bottom:14px;font-weight:700;color:#e6f6ff;">
    <div style="width:10px;height:10px;background:#7cf6a4;border-radius:4px;"></div>
    <div>TASK (c) – Compute the Surface Area as \(t \to \infty\)</div>
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    Substitute the antiderivative into the definite integral:
    \[
      S = 2\pi \lim_{t \to \infty} [\,I(t) - I(1)\,].
    \]
    As \(t \to \infty\):
    \[
      \operatorname{arsinh}(t^2) \sim \ln(2t^2) \to \infty, \quad
      \sqrt{1+\tfrac{1}{t^4}} \to 1.
    \]
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    Therefore, the limit diverges:
    \[
      S = 2\pi \times \infty = \boxed{\infty}.
    \]
  </div>
  <div style="color:#94a3b8;font-size:13px;">
    <strong>Result:</strong> Gabriel’s Horn has a finite volume \(V = \pi\), but an infinite surface area \(S = \infty\).
  </div>
</div>
</div>


<div>
    <p style="border-bottom: 2px solid black; border-top: 2px solid black; padding-bottom: 2em; padding-top: 2em; font-family: 'Courier New', Courier, monospace;">
    The reason why this is an inherent paradox can be explained in terms of paint. The volume of the horn (diverges) is able to be filled completely with paint. While the surface area (diverges) needs an inifinite amount of paint to coat it.</p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from scipy.special import gamma, factorial

x = np.linspace(1, 20, 500)
theta = np.linspace(0, 2 * np.pi, 300)
X, Theta = np.meshgrid(x, theta)
R = 1 / X
Y = R * np.cos(Theta)
Z = R * np.sin(Theta)
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')
norm = plt.Normalize(R.min(), R.max())
colors = cm.viridis(norm(R))

surf = ax.plot_surface(
    X, Y, Z,
    facecolors=colors,
    rstride=3,
    cstride=3,
    linewidth=0,
    antialiased=True,
    shade=True
)

mappable = cm.ScalarMappable(norm=norm, cmap=cm.viridis)
fig.colorbar(mappable, ax=ax, shrink=0.5, aspect=10, label=r"Radius $\left(\frac{1}{x}\right)$")
ax.set_title(r"Gabriel's Horn — $y = \frac{1}{x}$ Revolved About the x-axis", 
             pad=20, fontsize=14, weight='bold')

ax.set_xlabel(r"$x$", labelpad=10)
ax.set_ylabel(r"$y$", labelpad=10)
ax.set_zlabel(r"$z$", labelpad=10)
ax.set_xlim(1, 10)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
ax.set_box_aspect((3, 1, 1))
ax.view_init(elev=25, azim=-70)
fig.patch.set_facecolor("#f8f9fa")
ax.set_facecolor("#f0f0f0")
plt.tight_layout()
plt.show()
```

        The next question asks to consider the function $g(x)=\frac{x^a}{x!}$. We let $a=8$. Below I'm graphing the function for $0 \le x \le 20$. Since $x$ is discrete, $x!$ is only defined only for $x \in \mathbb{Z}_{\ge 0}$.
    </p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma, factorial

class FactorialPlot:
    def __init__(self, a=8, x_min=0, x_max=20, num_points=400):
        self.a = a
        self.x_min = x_min
        self.x_max = x_max
        self.num_points = num_points
        self.x_discrete = np.arange(self.x_min, self.x_max + 1)
        self.x_continuous = np.linspace(self.x_min, self.x_max, self.num_points)
        self.g_discrete = self.compute_discrete(self.x_discrete)
        self.g_continuous = self.compute_continuous(self.x_continuous)
        self.peak_x = self.x_discrete[np.argmax(self.g_discrete)]
        self.x_large = np.arange(50, 200, 10)
        self.g_values = self.x_large**a / factorial(self.x_large)
        self.styles = {
            "figsize": (10, 6),
            "continuous": {"color": "#1f77b4", "linewidth": 3, "alpha": 0.7, "label": "Gamma extension (continuous)"},
            "discrete": {"color": "#ff7f0e", "s": 100, "edgecolor": "k", "zorder": 5, "label": "Discrete factorial points"},
            "grid": {"which": "both", "linestyle": "--", "linewidth": 0.5, "alpha": 0.5},
            "title": {"fontsize": 16, "color": "#2ca02c"},
            "xlabel": {"fontsize": 14, "fontweight": "bold"},
            "ylabel": {"fontsize": 14, "fontweight": "bold"},
            "legend": {"frameon": True, "shadow": True, "fontsize": 12},
            "peak_line": {"color": "purple", "linestyle": ":", "linewidth": 2, "label": "Peak value"}
        }

    def compute_discrete(self, x):
        return x**self.a / factorial(x)

    def compute_continuous(self, x):
        return x**self.a / gamma(x + 1)
    
    def plot(self, title=None):
        plt.figure(figsize=self.styles["figsize"])
        plt.plot(self.x_continuous, self.g_continuous, **self.styles["continuous"])
        plt.scatter(self.x_discrete, self.g_discrete, **self.styles["discrete"])
        plt.axvline(x=self.peak_x, **self.styles["peak_line"])
        plt.grid(**self.styles["grid"])
        plt.xlabel("x", **self.styles["xlabel"])
        plt.ylabel("g(x)", **self.styles["ylabel"])
        if title is None:
            title = f"Plot of g(x)=x^{self.a} / x! for {self.x_min} ≤ x ≤ {self.x_max}"
        plt.title(title, **self.styles["title"])
        plt.legend(**self.styles["legend"])
        plt.minorticks_on()
        plt.tight_layout()
        plt.show()
    
    def lim_estimate(self):
        print(f"Estimating lim x→∞ g(x) = x^{self.a} / x!")
        for x, g in zip(self.x_large, self.g_values):
            print(f"x = {x:3}, g(x) = {g:.3e}")
        print("\nObservation: As x increases, g(x) approaches 0.")
        print(f"Estimated limit ≈ {self.g_values[-1]:.3e}")

    def formal_limit_N(self, epsilon=None):
        if epsilon is None:
            epsilon = 1 / self.a**3
        x = self.a
        while True:
            g = x**self.a / factorial(x)
            if g < epsilon:
                N = x
                break
            x += 1
        print(f"Formal ε-N limit check:")
        print(f"ε = {epsilon:.3e}")
        print(f"Smallest N such that |g(x)| < ε for all x > N: N = {N}")
        return N

plotter = FactorialPlot(a=8, x_max=20)
plotter.plot()
plotter.lim_estimate()
N = plotter.formal_limit_N()
```

        The final question asks to consider the series $$\sum_{n=1}^{\infty} \frac{(\ln(n))^2}{n^2}$$
    </p>
</div>
<div>
    <p style="font-family: 'Courier New', Courier, monospace; margin-top: 2em; margin-bottom: 2em; border-bottom: solid black 2px; padding-bottom: 2em;">
        The Python code below shows that the series is convergent, $S$. It also prints the error in $S \approx S_a$, $S \approx S_{a+100}$, and $S \approx S_{a+1000}$.
    </p>
</div>

```python
class LogSquaredSeries:
    def __init__(self, a):
        self.a = a
        self.n_values = np.arange(1, self.a + 1)
        self.S_a = self.compute_partial_sum()
        self.error_bound = self.compute_error_bound()
    
    def term(self, n):
        return (np.log(n))**2 / n**2
    
    def compute_partial_sum(self, n_max=None):
        if n_max is None:
            n_max = self.a
        n_values = np.arange(1, n_max + 1)
        return np.sum(self.term(n_values))
    
    def compute_error_bound(self, start=None):
        if start is None:
            start = self.a
        ln_start = np.log(start)
        return (ln_start**2 + 2*ln_start + 2) / start
    
    def summary(self):
        print(f"Partial sum S_{self.a} = {self.S_a:.6f}")
        print(f"Estimated error |S - S_{self.a}| ≤ {self.error_bound:.6f}")
    
    def error_for_offset(self, k_values):
        for k in k_values:
            n_new = self.a + k
            S_new = self.compute_partial_sum(n_new)
            error_new = self.compute_error_bound(n_new)
            print(f"S_{self.a}+{k} = S_{n_new} ≈ {S_new:.6f}, estimated error ≤ {error_new:.6f}")

series = LogSquaredSeries(a=10)
series.summary()
series.error_for_offset([100, 1000])
```

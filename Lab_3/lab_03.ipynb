  The paradox in question is that of Gabriel's/Torricelli's Horn; a function mapped such that it contains a finite volume but infinite surface area.

<p style="font-size: 1em; overflow-wrap: normal; font-family: 'Courier New', Courier, monospace;">
  I'm asked to consider the function $f(x)=\frac{1}{x}$ for $x>1$ and revolve it about the x-axis to get the illusive horn. Then, compute the finite volume of the horn for $1 < x < \infty$ and take the integral of the function for the surface area.

<div style="font-size:20px;margin-bottom:10px;color:#7dd3fc;">Partial Fraction Decomposition - Step by Step</div>
<div style="color:#94a3b8;font-size:13px;margin-bottom:18px;">Rational integrand</div>
<div style="background:rgba(255,255,255,0.02);border:1px solid rgba(125,211,252,0.06);border-radius:12px;padding:20px;margin-bottom:24px;box-shadow:0 6px 30px rgba(2,6,23,0.6);">
  <div style="display:flex;align-items:center;gap:12px;margin-bottom:14px;font-weight:700;color:#e6f6ff;">
    <div style="width:10px;height:10px;background:#7dd3fc;border-radius:4px;"></div>
    <div>STEP 1 - Factor the denominator</div>
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    \( \displaystyle
      \text{Integrand: }
      \frac{4x^{3}-27x^{2}+5x-32}{30x^{5}-13x^{4}+50x^{3}-286x^{2}-299x-70}
    \)
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    Factor the denominator:
    \[
      30x^{5}-13x^{4}+50x^{3}-286x^{2}-299x-70
      = (2x+1)(3x-7)(5x+2)(x^{2}+x+5)
    \]
  </div>
  <div style="color:#94a3b8;font-size:13px;">
    So now decompose with linear factors \(2x+1,\,3x-7,\,5x+2\) and the irreducible quadratic \(x^{2}+x+5\).
  </div>
</div>
<div style="background:rgba(255,255,255,0.02);border:1px solid rgba(125,211,252,0.06);border-radius:12px;padding:20px;margin-bottom:24px;box-shadow:0 6px 30px rgba(2,6,23,0.6);">
  <div style="display:flex;align-items:center;gap:12px;margin-bottom:14px;font-weight:700;color:#e6f6ff;">
    <div style="width:10px;height:10px;background:#ffd166;border-radius:4px;"></div>
    <div>STEP 2 - Partial fraction setup and solved coefficients</div>
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    Set up the decomposition:
    \[
      \frac{4x^{3}-27x^{2}+5x-32}{(2x+1)(3x-7)(5x+2)(x^{2}+x+5)}
      = \frac{A}{2x+1}+\frac{B}{3x-7}+\frac{C}{5x+2}+\frac{Dx+E}{x^{2}+x+5}
    \]
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    Solving for constants:
  </div>
  <div style="display:grid;grid-template-columns:repeat(auto-fit,minmax(220px,1fr));gap:2em;margin-top:1em;margin-bottom:1em;">
    <div style="background:rgba(255,255,255,0.02);padding:1em;border-radius:8px;border:1px solid rgba(125,211,252,0.04);">
      \(\displaystyle A=-\frac{668}{323}\)
      <div style="color:#94a3b8;font-size:13px;margin-top:8px;">term: \( \frac{A}{2x+1}\)</div>
    </div>
    <div style="background:rgba(255,255,255,0.02);padding:1em;border-radius:8px;border:1px solid rgba(125,211,252,0.04);">
      \(\displaystyle B=-\frac{9438}{80155}\)
      <div style="color:#94a3b8;font-size:13px;margin-top:8px;">term: \( \frac{B}{3x-7}\)</div>
    </div>
    <div style="background:rgba(255,255,255,0.02);padding:1em;border-radius:8px;border:1px solid rgba(125,211,252,0.04);">
      \(\displaystyle C=\frac{24110}{4879}\)
      <div style="color:#94a3b8;font-size:13px;margin-top:8px;">term: \( \frac{C}{5x+2}\)</div>
    </div>
    <div style="background:rgba(255,255,255,0.02);padding:1em;border-radius:8px;border:1px solid rgba(125,211,252,0.04);">
      \(\displaystyle D=\frac{22098}{260015}\)
      <div style="color:#94a3b8;font-size:13px;margin-top:8px;">quadratic numerator coefficient of \(x\)</div>
    </div>
    <div style="background:rgba(255,255,255,0.02);padding:1em;border-radius:8px;border:1px solid rgba(125,211,252,0.04);">
      \(\displaystyle E=\frac{9787}{52003}\)
      <div style="color:#94a3b8;font-size:13px;margin-top:8px;">quadratic numerator constant</div>
    </div>
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    So the decomposition is:
    \[
      \frac{4x^{3}-27x^{2}+5x-32}{\dots}
      = -\frac{668}{323(2x+1)}
        -\frac{9438}{80155(3x-7)}
        +\frac{24110}{4879(5x+2)}
        +\frac{\tfrac{22098}{260015}x+\tfrac{9787}{52003}}{x^{2}+x+5}
    \]
  </div>
</div>
<div style="background:rgba(255,255,255,0.02);border:1px solid rgba(125,211,252,0.06);border-radius:12px;padding:20px;margin-bottom:24px;box-shadow:0 6px 30px rgba(2,6,23,0.6);">
  <div style="display:flex;align-items:center;gap:12px;margin-bottom:14px;font-weight:700;color:#e6f6ff;">
    <div style="width:10px;height:10px;background:#7cf6a4;border-radius:4px;"></div>
    <div>STEP 3 - Integrate each term</div>
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    Integrate term-by-term:
    \[
      \int \frac{A}{2x+1}\,dx=\frac{A}{2}\ln|2x+1|,\quad
      \int\frac{B}{3x-7}\,dx=\frac{B}{3}\ln|3x-7|,\quad
      \int\frac{C}{5x+2}\,dx=\frac{C}{5}\ln|5x+2|
    \]
  </div>
  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
    For the quadratic term write \(Dx+E=\tfrac{D}{2}(2x+1)+(E-\tfrac{D}{2})\):
    \[
      \int\frac{Dx+E}{x^{2}+x+5}\,dx
      = \frac{D}{2}\ln(x^{2}+x+5)
        +(E-\tfrac{D}{2})\int\frac{dx}{x^{2}+x+5}
    \]
    Completing the square:
    \(x^{2}+x+5=(x+\tfrac12)^{2}+\tfrac{19}{4}\), so
    \[
      \int\frac{dx}{x^{2}+x+5}
      = \frac{2}{\sqrt{19}}\arctan\!\Big(\frac{2x+1}{\sqrt{19}}\Big)
    \]
  </div>
  <div style="background:linear-gradient(180deg,rgba(34,197,94,0.03),rgba(14,42,20,0.02));border:1px solid rgba(34,197,94,0.06);border-radius:8px;padding:16px;margin-top:20px;">
    <strong>Final antiderivative:</strong>
    \[
      \boxed{\,\begin{aligned}
        \int &\frac{4x^{3}-27x^{2}+5x-32}{30x^{5}-13x^{4}+50x^{3}-286x^{2}-299x-70}\,dx \;=\\[6pt]
        &-\frac{334}{323}\ln|2x+1|
        -\frac{3146}{80155}\ln|3x-7|
        +\frac{4822}{4879}\ln|5x+2|\\[6pt]
        &\quad+\frac{11049}{260015}\ln(x^{2}+x+5)
        +\frac{3988\sqrt{19}}{260015}\arctan\!\Big(\frac{2x+1}{\sqrt{19}}\Big)+C.
      \end{aligned}}
    \]
  </div>
  <div style="color:#94a3b8;font-size:13px;margin-top:12px;">
    Equivalent rationalized coefficient form:
    \(\frac{3988}{13685\sqrt{19}}=\frac{3988\sqrt{19}}{260015}\).
  </div>
</div>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

x = np.linspace(1, 20, 500)
theta = np.linspace(0, 2 * np.pi, 300)
X, Theta = np.meshgrid(x, theta)
R = 1 / X
Y = R * np.cos(Theta)
Z = R * np.sin(Theta)
fig = plt.figure(figsize=(12, 7))
ax = fig.add_subplot(111, projection='3d')
norm = plt.Normalize(R.min(), R.max())
colors = cm.viridis(norm(R))

surf = ax.plot_surface(
    X, Y, Z,
    facecolors=colors,
    rstride=3,
    cstride=3,
    linewidth=0,
    antialiased=True,
    shade=True
)

mappable = cm.ScalarMappable(norm=norm, cmap=cm.viridis)
fig.colorbar(mappable, ax=ax, shrink=0.5, aspect=10, label=r"Radius $\left(\frac{1}{x}\right)$")
ax.set_title(r"Gabriel's Horn  $y = \frac{1}{x}$ Revolved About the x-axis", 
             pad=20, fontsize=14, weight='bold')

ax.set_xlabel(r"$x$", labelpad=10)
ax.set_ylabel(r"$y$", labelpad=10)
ax.set_zlabel(r"$z$", labelpad=10)
ax.set_xlim(1, 10)
ax.set_ylim(-1, 1)
ax.set_zlim(-1, 1)
ax.set_box_aspect((3, 1, 1))
ax.view_init(elev=25, azim=-70)
fig.patch.set_facecolor("#f8f9fa")
ax.set_facecolor("#f0f0f0")
plt.tight_layout()
plt.show()
```

<div>
    <p style="font-family: 'Courier New', Courier, monospace;">
        Using that equation, I'll compute
    </p>
</div>
<div style="font-family: 'Courier New', Courier, monospace;">
    \[
        \int_{1}^{t} g(x) \, dx    
    \]
</div>
<div style="font-family: 'Courier New', Courier, monospace; margin-top: 2em;">
    for $t=2, 5, 10, 100, 1000, 10000$.
</div>
<div style="font-family: 'Courier New', Courier, monospace; margin-top: 2em;">
    For which, as $t$ appraches $\infty$ the function converges since the integrand is not negative.
</div>
<div>
    \[
        \lim_{t\rightarrow \infty} \int_{1}^{t} \frac{\cos^2(x)}{x^2}dx = \frac{1}{2}(1+\cos(2))-\frac{\pi}{2}+Si(2)
    \]
</div>

```python
from mpmath import mp, cos, si
import matplotlib.pyplot as plt
mp.dps = 50
def I(t):
    return 0.5*(1 + cos(2)) - (1/(2*t))*(1 + cos(2*t)) - (si(2*t) - si(2))
t_values = [2, 5, 10, 100, 1000, 10000]
I_values = [float(I(t)) for t in t_values]
for t, val in zip(t_values, I_values):
    print(f"t = {t:<6}  =>  I(t) = {val:.12f}")
plt.figure(figsize=(8, 5))
plt.plot(t_values, I_values, marker='o', linewidth=1.8)
plt.xscale('log')
plt.title(r"$I(t)=\int_1^t \frac{\cos^2(x)}{x^2}\,dx$", fontsize=14)
plt.xlabel("t (log scale)")
plt.ylabel("I(t)")
plt.grid(True, which="both", linestyle="--", alpha=0.6)
plt.show()
```

```python
import numpy as np
f = lambda x: 1 / x**2
g = lambda x: 1 / x**3
x = np.linspace(1, 10, 500)
plt.figure(figsize=(8,5))
plt.plot(x, f(x), label=r'$f(x)=\frac{1}{x^2}$', color='blue', linewidth=2)
plt.plot(x, g(x), label=r'$g(x)=\frac{1}{x^3}$', color='red', linestyle='--', linewidth=2)
plt.fill_between(x, g(x), color='red', alpha=0.2)
plt.title(r'Comparison of $f(x)=\frac{1}{x^2}$ and $g(x)=\frac{1}{x^3}$')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0, 1)
plt.legend()
plt.grid(True)
plt.show()
```

        The third and final question asks to consider the curve $y=\log_ax$, where $a$ is the first initial of my first name (S). (if $a=1$, please use $a=e$, Euler's number)
    </p>
</div>
<div>
    <p style="font-family: 'Courier New', Courier, monospace; margin-top: 2em;">
        Below shows the curve sketch from the points (1, 0) to (2, $\log_a2$). It also computes the arc length between the points.
    </p>
</div>

```python
import math

initial = 'S'
a_raw = ord(initial.upper()) - ord('A') + 1
a = math.e if a_raw == 1 else float(a_raw)
def log_a(x, a):
    return np.log(x) / np.log(a)

def arc_length_exact(a):
    ln_a = math.log(a)
    k = 1.0 / ln_a
    def F(x):
        return math.sqrt(x*x + k*k) - k * math.asinh(k / x)
    return F(2.0) - F(1.0)

def arc_length_numeric(a, n=200000):
    xs = np.linspace(1.0, 2.0, n)
    dx = xs[1] - xs[0]
    integrand = np.sqrt(1.0 + (1.0 / (xs * math.log(a)))**2)
    return np.sum((integrand[:-1] + integrand[1:]) / 2 * dx)

L_exact = arc_length_exact(a)
L_num = arc_length_numeric(a, n=20000)
print(f"Initial: '{initial}', raw index = {a_raw}")
print(f"Using base a = {a:.12g}")
print(f"Exact arc length from x=1 to x=2: L = {L_exact:.12f}")
print(f"Numeric trapezoid check: L \u2248 {L_num:.12f}")
x_plot = np.linspace(1.0, 2.0, 1000)
y_plot = log_a(x_plot, a)

plt.figure(figsize=(8,5))
plt.plot(x_plot, y_plot, label=rf"$y=\log_{{{a:.0f}}} x$" if a==int(a) else rf"$y=\log_{{{a:.6g}}} x$", linewidth=2)
x_pts = np.array([1.0, 2.0])
y_pts = np.array([0.0, np.log(2.0)/np.log(a)])
plt.scatter(x_pts, y_pts, color='red', zorder=5)
plt.plot(x_pts, y_pts, color='gray', linestyle='--', alpha=0.7, label='Chord between endpoints')
plt.annotate(f"L = {L_exact:.6f}", xy=(1.5, (y_pts[0]+y_pts[1])/2),
             xytext=(1.05, max(y_plot)*0.6),
             arrowprops=dict(arrowstyle="->", lw=1), fontsize=10, bbox=dict(boxstyle="round", fc="w"))
plt.xlim(0.95, 2.05)
ymin = min(y_plot.min(), -0.05)
ymax = max(y_plot.max(), y_pts.max())*1.08
plt.ylim(ymin, ymax)
plt.xlabel("x")
plt.ylabel("y")
plt.title(f"Plot of y = log_{a:.6g}(x) from x=1 to x=2")
plt.legend()
plt.grid(alpha=0.35)
plt.show()
```

  <div style="font-size:20px;margin-bottom:10px;color:#7dd3fc;">Bessel Function of the First Kind</div>
  <div style="color:#94a3b8;font-size:13px;margin-bottom:18px;">Series convergence and partial-sum approximations for \(a = 8\)</div>

  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:24px;border:1px solid rgba(255,255,255,0.06);">
    Let \(a = 8\). Consider the Bessel function of the first kind
    \[
      J_a(x) = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n+a}}{n!(n+a)!\,2^{2n+a}}.
    \]
    For \(a = 8\) this becomes
    \[
      J_8(x) = \sum_{n=0}^{\infty} \frac{(-1)^n x^{2n+8}}{n!(n+8)!\,2^{2n+8}}.
    \]
  </div>

  <!-- PART (a) -->
  <div style="background:rgba(255,255,255,0.02);border:1px solid rgba(125,211,252,0.06);border-radius:12px;padding:20px;margin-bottom:24px;box-shadow:0 6px 30px rgba(2,6,23,0.6);">
    <div style="display:flex;align-items:center;gap:12px;margin-bottom:14px;font-weight:700;color:#e6f6ff;">
      <div style="width:10px;height:10px;background:#7dd3fc;border-radius:4px;"></div>
      <div>TASK (a) – Interval of Convergence via the Ratio Test</div>
    </div>

  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
      We write the series as
      \[
        J_8(x) = \sum_{n=0}^{\infty} a_n(x),
        \qquad
        a_n(x) = \frac{(-1)^n x^{2n+8}}{n!(n+8)!\,2^{2n+8}}.
      \]
      To find the interval of convergence, we apply the Ratio Test to the general term \(a_n(x)\).
  </div>

  <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
      We consider
      \[
        \left|\frac{a_{n+1}(x)}{a_n(x)}\right|
        = \left|
          \frac{(-1)^{n+1} x^{2(n+1)+8}}{(n+1)!(n+9)!\,2^{2(n+1)+8}}
          \cdot
          \frac{n!(n+8)!\,2^{2n+8}}{(-1)^n x^{2n+8}}
        \right|.
      \]
      Simplifying step by step:
      \[
        x^{2(n+1)+8} = x^{2n+10} = x^{2n+8}\,x^2,
      \]
      \[
        \frac{n!}{(n+1)!} = \frac{1}{n+1}, \qquad
        \frac{(n+8)!}{(n+9)!} = \frac{1}{n+9},
      \]
      \[
        \frac{2^{2n+8}}{2^{2n+10}} = \frac{1}{2^2} = \frac{1}{4}.
      \]
      Therefore
      \[
        \left|\frac{a_{n+1}(x)}{a_n(x)}\right|
        = |x|^2 \cdot \frac{1}{n+1} \cdot \frac{1}{n+9} \cdot \frac{1}{4}
        = \frac{|x|^2}{4(n+1)(n+9)}.
      \]
      Now take the limit as \(n \to \infty\):
      \[
        \lim_{n \to \infty}
        \left|\frac{a_{n+1}(x)}{a_n(x)}\right|
        = \lim_{n \to \infty} \frac{|x|^2}{4(n+1)(n+9)} = 0
      \]
      for every real number \(x\). Since \(0 &lt; 1\), the Ratio Test tells us that the series
      \(\sum_{n=0}^{\infty} a_n(x)\) converges (in fact, absolutely) for all real \(x\).
  </div>

  <div style="color:#94a3b8;font-size:13px;">
      <strong>Conclusion:</strong> The power series for \(J_8(x)\) converges for every real number \(x\).<br>
      The interval of convergence is
      \[
        (-\infty, \infty),
      \]
      so the radius of convergence is \(R = \infty\).
  </div>
  </div>

  <div style="background:rgba(255,255,255,0.02);border:1px solid rgba(125,211,252,0.06);border-radius:12px;padding:20px;margin-bottom:8px;box-shadow:0 6px 30px rgba(2,6,23,0.6);">

      <div style="width:10px;height:10px;background:#f97373;border-radius:4px;"></div>
      <div>TASK (d) – Describe the Quality of the Approximation</div>
    </div>
    <div style="background:#071025;border-radius:8px;padding:16px;font-family:'Courier New',monospace;white-space:pre-wrap;margin-bottom:18px;border:1px solid rgba(255,255,255,0.03);">
      For each integer \(b \ge 0\), the partial sum
      \[
        S_b(x) = \sum_{n=0}^{b} \frac{(-1)^n x^{2n+8}}{n!(n+8)!\,2^{2n+8}}
      \]
      approximates the full Bessel function \(J_8(x)\). Since the series has infinite radius of convergence, we have
      \[
        \lim_{b \to \infty} S_b(x) = J_8(x)
      \]
      for every fixed real \(x\). Thus the approximation improves as \(b\) increases.
      <br><br>
      Near \(x = 0\), the higher powers \(x^{2n+8}\) are very small, so only a few terms are needed for a good approximation: even small values of \(b\) give \(S_b(x)\) very close to \(J_8(x)\) near the origin. For larger values of \(x\), more terms are required before the alternating series and factorials in the denominator dominate; as \(b\) increases, the partial sums capture more of the oscillatory behavior of \(J_8(x)\), and the amplitudes and zeros of \(S_b(x)\) line up more closely with those of the full function.
    </div>
    <div style="color:#94a3b8;font-size:13px;">
      In summary, the partial sums \(S_b(x)\) converge to \(J_8(x)\) for all real \(x\), with especially rapid convergence near \(x = 0\) and increasingly accurate approximation on \(0 \le x \le 10\) as the truncation level \(b\) grows.
    </div>
  </div>
</div>


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # registers 3D projection
from scipy.special import jv
import math
import sympy as sp
from sympy.plotting import plot_parametric
import numpy as np
import mpmath as mp


a = 8
x_min, x_max = 0.0, 10.0
num_x = 400
b_max = 50

x = np.linspace(x_min, x_max, num_x)

def bessel_partial_sum(a, x, b):
    x = np.asarray(x, dtype=float)
    S = np.zeros_like(x, dtype=float)

    for n in range(b + 1):
        coef = ((-1)**n) / (math.factorial(n) * math.factorial(n + a) * 2.0**(2*n + a))
        S += coef * x**(2*n + a)

    return S


S_rows = []
for b in range(b_max + 1):
    S_rows.append(bessel_partial_sum(a, x, b))

S = np.vstack(S_rows)

B_indices = np.arange(0, b_max + 1)
X, B = np.meshgrid(x, B_indices)
J_exact = jv(a, x)
fig = plt.figure(figsize=(11, 7))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(
    X, B, S,
    cmap='viridis',
    linewidth=0,
    antialiased=True,
    alpha=0.85
)

highlight_bs = [1, 10, 50]
for b in highlight_bs:
    ax.plot(
        x,
        np.full_like(x, b, dtype=float),
        S[b],
        linewidth=2,
        label=fr'$S_{b}(x)$'
    )

b_exact_level = b_max + 5
ax.plot(
    x,
    np.full_like(x, b_exact_level, dtype=float),
    J_exact,
    color='red',
    linewidth=3,
    label=r'$J_8(x)$'
)

ax.set_xlabel(r'$x$', labelpad=10)
ax.set_ylabel(r'Number of terms $b$', labelpad=10)
ax.set_zlabel(r'$S_b(x)$ or $J_8(x)$', labelpad=10)
ax.set_title(r'3D View of Partial Sums $S_b(x)$ Approaching $J_8(x)$', pad=15)
cbar = fig.colorbar(surf, ax=ax, shrink=0.6, aspect=14)
cbar.set_label(r'$S_b(x)$', rotation=90, labelpad=10)
ax.view_init(elev=30, azim=-60)
ax.legend(loc='upper left', bbox_to_anchor=(1.02, 1.0))

plt.tight_layout()
plt.show()
```

        The next question asks to consider the Lissajous figure. I'll use the same definition of $a=8$ and $b$ from above. The variable $t$ though, varies between 0 and $2\pi$.
        $$\Biggl\{ \begin{matrix*}
        x=a\sin{(3t)}\\
        y=b\cos{(t)}
        \end{matrix*}$$
    </p>
</div>

```python
t = sp.symbols('t', real=True)
x_var = sp.symbols('x_var')

a = 8
b = 10

x_t = a * sp.sin(3*t)
y_t = b * sp.cos(t)
dxdt = sp.diff(x_t, t)
dydt = sp.diff(y_t, t)

curve_plot = plot_parametric(
    (x_t, y_t, (t, 0, 2*sp.pi)),
    xlabel='x',
    ylabel='y',
    title='Lissajous Curve: x = a sin(3t), y = b cos(t)',
    show=False
)
curve_plot.show()
t0 = sp.pi/12
slope = sp.simplify(dydt / dxdt)
x0 = sp.simplify(x_t.subs(t, t0))
y0 = sp.simplify(y_t.subs(t, t0))
m0 = sp.simplify(slope.subs(t, t0))

print("=== Tangent at t = π/12 ===")
print("x0 =", x0)
print("y0 =", y0)
print("slope m0 =", m0)

tangent_line_expr = sp.simplify(m0 * (x_var - x0) + y0)
print("Tangent line equation in y(x):")
print("y =", tangent_line_expr)

area_integrand = y_t * dxdt
f_area = sp.lambdify(t, area_integrand, 'mpmath')

mp.mp.dps = 30

area_numeric = mp.quad(f_area, [0, 2*mp.pi])

print("\n=== Area under the curve (numeric) ===")
print("A ≈", area_numeric)

length_integrand = sp.sqrt(dxdt**2 + dydt**2)
f_length = sp.lambdify(t, length_integrand, 'mpmath')

length_numeric = mp.quad(f_length, [0, 2*mp.pi])

print("\n=== Arc length of the curve (numeric) ===")
print("L ≈", length_numeric)

t_vals = np.linspace(0, 2*np.pi, 1000)
x_vals = a * np.sin(3*t_vals)
y_vals = b * np.cos(t_vals)

x0_float = float(x0.evalf())
y0_float = float(y0.evalf())
m0_float = float(m0.evalf())

u_vals = np.linspace(-5, 5, 200)
x_tan_vals = x0_float + u_vals
y_tan_vals = y0_float + m0_float * u_vals
z_vals = t_vals

fig = plt.figure(figsize=(12, 5))

ax2d = fig.add_subplot(1, 2, 1)
ax2d.plot(x_vals, y_vals, label='Lissajous curve')
ax2d.plot(x_tan_vals, y_tan_vals, 'r--', label='Tangent at t = π/12')
ax2d.scatter([x0_float], [y0_float], color='k', zorder=5, label='Point at t = π/12')
ax2d.set_xlabel('x')
ax2d.set_ylabel('y')
ax2d.set_title('2D Lissajous Curve with Tangent')
ax2d.legend()
ax2d.grid(True)

ax3d = fig.add_subplot(1, 2, 2, projection='3d')
ax3d.plot(x_vals, y_vals, z_vals, label='(x(t), y(t), t)')
ax3d.set_xlabel('x')
ax3d.set_ylabel('y')
ax3d.set_zlabel('t')
ax3d.set_title('3D View of Lissajous Curve (t as height)')
ax3d.legend()

plt.tight_layout()
plt.show()
```
